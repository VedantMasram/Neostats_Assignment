# Q7. Which age group spends the most?

```python
j=df
```

```python
j.Age.unique()
```

```python
def age_groups(Age):
    if Age >=18 and Age <=30:
        return '18-30'
    elif Age >=31 and Age <=40:
        return '31-40'
    elif Age >=41 and Age<= 50:
        return '41-50'
    elif Age>=51 and Age<=60:
        return '51-60'
    elif Age>=61 and Age<= 70:
        return '61-70'
    else:
        return '71-80'
```

```python
j['AgeGroups']=j['Age'].apply(age_groups)
```

```python
j
```

```python
ss.groupby('AgeGroups')['Customer'].count()
```

```python
highest_spent=j.groupby('AgeGroups')['Amount'].sum()
highest_spent
```

```python
highest_spent.plot(kind='bar')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=0)
plt.show()
```
